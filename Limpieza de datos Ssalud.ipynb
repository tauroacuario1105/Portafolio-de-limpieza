```python
# Convertir columnas de tipo int64 a datetime
df['FECHA_CORTE'] = pd.to_datetime(df['FECHA_CORTE'], format='%Y%m%d', errors='coerce')
df['FECHA_MUESTRA'] = pd.to_datetime(df['FECHA_MUESTRA'], format='%Y%m%d', errors='coerce')
df['FEC_RESULTADO_1'] = pd.to_datetime(df['FEC_RESULTADO_1'], format='%Y%m%d', errors='coerce')
df['FEC_RESULTADO_2'] = pd.to_datetime(df['FEC_RESULTADO_2'], format='%Y%m%d', errors='coerce')

# Mostrar los tipos de datos actualizados
print(df.dtypes)
```

```python
df
```

```python


metricas = df['DEPARTAMENTO'].value_counts()

plt.figure(figsize=(10, 6))
metricas.plot(kind='bar', color='skyblue')
plt.title('Número de Pacientes por Departamento')
plt.xlabel('Departamento')
plt.ylabel('Cantidad de Pacientes')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python
# Asegúrate de que la columna de fechas está en formato datetime
df['FECHA_MUESTRA'] = pd.to_datetime(df['FECHA_MUESTRA'])

# Crear una nueva columna con el mes (usando Periodo mensual)
df['MES'] = df['FECHA_MUESTRA'].dt.to_period('M')

# Contar la cantidad de procedimientos realizados por mes
metricas = df.groupby('MES').size()

# Graficar los resultados
plt.figure(figsize=(12, 6))
metricas.plot(kind='line', marker='o', color='green')
plt.title('Cantidad de Procedimientos Realizados por Mes')
plt.xlabel('Mes')
plt.ylabel('Cantidad de Procedimientos')
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()

```
